Notebook de exploracao inicial dos dados sinteticos de pacientes, comorbidades e prescricoes.

**Pre-requisito:** rodar `scripts/generate_synthetic_data.py` para popular `data/raw/`.

Aviso: dados 100% sinteticos — uso educacional/pesquisa.

```python
import pandas as pd

patients = pd.read_csv("../data/raw/patients.csv")
comorbidities = pd.read_csv("../data/raw/comorbidities.csv")
prescriptions = pd.read_csv("../data/raw/prescriptions.csv")

patients.describe(include="all")
```

```python
# Distribuicao do numero de medicamentos ativos por paciente
active_rx = prescriptions[prescriptions["end_date"].isna()]
med_counts = active_rx.groupby("patient_id").size()
med_counts.describe()
```

```python
import matplotlib.pyplot as plt

med_counts.plot(kind="hist", bins=15, title="Medicamentos ativos por paciente")
plt.axvline(5, color="orange", linestyle="--", label="Limiar polifarmacia")
plt.axvline(10, color="red", linestyle="--", label="Limiar hiperpolifarmacia")
plt.legend()
plt.show()
```
